````
import os
import pandas as pd

def read_all_data(base_path):
    all_data = []

    for month in os.listdir(base_path):
        month_path = os.path.join(base_path, month)
        if not os.path.isdir(month_path):
            continue

        for file in os.listdir(month_path):
            if 'stop-and-search' not in file:
                continue

            file_path = os.path.join(month_path, file)

            try:
                df = pd.read_csv(file_path)

                # แยกชื่อเมืองจากชื่อไฟล์ เช่น avon-and-somerset
                city = file.replace('-stop-and-search.csv', '').split('-', 2)[-1].strip()
                df['เมือง'] = city
                all_data.append(df)
            except Exception as e:
                print(f"Error reading {file_path}: {e}")

    return pd.concat(all_data, ignore_index=True)
````

````
# 1. อ่านข้อมูลทั้งหมด
base_path = "/content/drive/MyDrive/Python&DataViz2025/data"
df_all = read_all_data(base_path)

# 2. กรองเฉพาะคนที่ระบุว่าเป็นคน Asia
# ใช้ self-defined-ethnicity ถ้ามี
eth_col = "Self-defined ethnicity" if "Self-defined ethnicity" in df_all.columns else "ethnicity"

df_asian = df_all[df_all[eth_col].str.contains("Asian", case=False, na=False)]

# 3. Group by เมืองแล้วนับจำนวน
summary = df_asian.groupby("เมือง").size().reset_index(name="Asian Count")

# 4. หาค่าน้อยสุดและมากสุด
min_row = summary.loc[summary["Asian Count"].idxmin()]
max_row = summary.loc[summary["Asian Count"].idxmax()]

print("เมืองที่มีคนเอเชียน้อยที่สุด:")
print(min_row)

print("\nเมืองที่มีคนเอเชียมากที่สุด:")
print(max_row)
````

#HW 5 (เดี่ยว) ***แก้จาก code เดิม*** ยอดรวมทุกเมือง เดือนไหนมีจำนวนผู้ถูกเรียกค้นที่เป็นคน Asia น้อยที่สุด และ มากที่สุด

````
import os
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd

def read_all_data(base_path):
    all_data = []

    for month in os.listdir(base_path):
        month_path = os.path.join(base_path, month)
        if not os.path.isdir(month_path):
            continue

        for file in os.listdir(month_path):
            if 'stop-and-search' not in file:
                continue

            file_path = os.path.join(month_path, file)

            try:
                df = pd.read_csv(file_path)

                # แยกชื่อเมืองจากชื่อไฟล์ เช่น avon-and-somerset
                city = file.replace('-stop-and-search.csv', '').split('-', 2)[-1].strip()
                df['เมือง'] = city
                all_data.append(df)
            except Exception as e:
                print(f"Error reading {file_path}: {e}")

    return pd.concat(all_data, ignore_index=True)
````

````
# 1. โหลดข้อมูล
path = "/content/drive/MyDrive/Python&DataViz2025/data"
df = read_all_data(path)

# 2. กรองเฉพาะคน Asian ใช้คอลัมม์ Officer-defined ethnicity เพราะคอลัมม์ Self-defined ethnicity บาง row ไม่มีคำว่า Asian
df = df[df["Officer-defined ethnicity"].str.contains("Asian", case=False, na=False)]

# 3. แปลงวัน และเพิ่มคอลัมน์เดือน
df["Date"] = pd.to_datetime(df["Date"], errors="coerce").dt.tz_localize(None)
df["Month"] = df["Date"].dt.to_period("M")

# 4. สรุปจำนวนในแต่ละเดือน
counts = df["Month"].value_counts().sort_index().reset_index()
counts.columns = ["Month", "Asian Count"]

# 5. แสดงเดือนน้อยสุดและมากสุด
print("เดือนที่มีจำนวนผู้ถูกเรียกค้นที่เป็นคน Asian น้อยที่สุด:\n",
      counts.nsmallest(1, "Asian Count").iloc[0].to_string())

print("\nเดือนที่มีจำนวนผู้ถูกเรียกค้นที่เป็นคน Asian มากที่สุด:\n",
      counts.nlargest(1, "Asian Count",).iloc[0].to_string())

````

````
# prompt: จาก data this_data วนลูปนับว่ามีนิทานที่มีคำว่า แม่ หรือ พ่อ ในชื่อเรื่อง ทั้งหมดกี่เรื่อง เรื่องอะไรบ้าง

count = 0
stories = []
for index, row in this_data.iterrows():
  if 'แม่' in row['title'] or 'พ่อ' in row['title']:
    count += 1
    stories.append(row['title'])

print(f"There are {count} stories with 'แม่' or 'พ่อ' in the title.")
print("The stories are:")
for story in stories:
  print(story)
````

````
# prompt: มีคำดังกล่าวใน นิทานชีวิต หรือ นิทานประจำถิ่นมากกว่ากัน บอกจำนวนด้วย

import pandas
import pandas as pd
import os
from google.colab import drive
from bs4 import BeautifulSoup

# ... (preceding code)

count_life = 0
count_local = 0
for index, row in this_data.iterrows():
  if ('แม่' in row['title'] or 'พ่อ' in row['title']) and row['type'] == 'นิทานชีวิต':
    count_life += 1
  elif ('แม่' in row['title'] or 'พ่อ' in row['title']) and row['type'] == 'นิทานประจำถิ่น':
    count_local += 1

if count_life > count_local:
  print(f"There are more stories with 'แม่' or 'พ่อ' in the title in นิทานชีวิต ({count_life} stories) than in นิทานประจำถิ่น ({count_local} stories).")
elif count_local > count_life:
  print(f"There are more stories with 'แม่' or 'พ่อ' in the title in นิทานประจำถิ่น ({count_local} stories) than in นิทานชีวิต ({count_life} stories).")
else:
  print(f"There are an equal number of stories with 'แม่' or 'พ่อ' in the title in both นิทานชีวิต and นิทานประจำถิ่น ({count_life} stories each).")
````

````
# prompt: คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่อง ว่า แม่ หรือ พ่อ มากที่สุด (พร้อมทั้งอธิบาย code แต่ละบรรทัดด้วย)

# สร้าง dictionary เพื่อเก็บจำนวนนิทานที่มีคำว่า 'แม่' หรือ 'พ่อ' ในชื่อเรื่องสำหรับแต่ละคนแต่ง
creator_counts = {}
for index, row in this_data.iterrows():
  if 'แม่' in row['title'] or 'พ่อ' in row['title']:
    creator = row['creator']
    if creator not in creator_counts:
      creator_counts[creator] = 0
    creator_counts[creator] += 1

# หาคนแต่งที่มีจำนวนนิทานมากที่สุด
max_count = 0
most_prolific_creator = None
for creator, count in creator_counts.items():
  if count > max_count:
    max_count = count
    most_prolific_creator = creator

print(f"The creator who wrote the most stories with 'แม่' or 'พ่อ' in the title is {most_prolific_creator} with {max_count} stories.")
````
